```python

# ===========================================================
# 1. INSTALL DAN IMPORT LIBRARY
# ===========================================================

!pip install mlxtend openpyxl -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from google.colab import files
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Library berhasil di-import.")


# ===========================================================
# 2. BUSINESS UNDERSTANDING
# ===========================================================

print("="*80)
print("BUSINESS UNDERSTANDING")
print("="*80)

print("""
Penelitian ini bertujuan untuk menemukan pola asosiasi dalam perilaku pemesanan
pelanggan hotel menggunakan algoritma Apriori pada dataset Hotel Booking Demand.

Masalah utama:
Hotel memiliki data pemesanan pelanggan dalam jumlah besar, tetapi pola hubungan
antar atribut pemesanan belum terlihat secara langsung.

Tujuan analisis:
1. Menemukan pola pemesanan pelanggan yang sering muncul.
2. Menemukan kombinasi atribut yang berkaitan dengan pembatalan booking.
3. Menemukan kombinasi atribut yang berkaitan dengan booking tidak dibatalkan.
4. Mengevaluasi aturan asosiasi menggunakan support, confidence, dan lift.

Catatan:
Pada Apriori, evaluasi tidak menggunakan accuracy seperti klasifikasi.
Nilai confidence digunakan sebagai pendekatan tingkat ketepatan aturan asosiasi.
""")


# ===========================================================
# 3. UPLOAD DAN LOAD DATASET
# ===========================================================

print("="*80)
print("UPLOAD DATASET")
print("="*80)

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
print("File yang digunakan:", file_name)

df = pd.read_csv(file_name)

print("\nJumlah baris dan kolom awal:", df.shape)
display(df.head())


# ===========================================================
# 4. DATA UNDERSTANDING
# ===========================================================

print("="*80)
print("DATA UNDERSTANDING")
print("="*80)

print("\nInformasi dataset:")
df.info()

print("\nDaftar kolom dataset:")
print(df.columns.tolist())

print("\nJumlah missing value per kolom:")
missing_before = df.isnull().sum().sort_values(ascending=False)
display(missing_before)

print("\nJumlah data duplikat:")
duplicate_before = df.duplicated().sum()
print(duplicate_before)

print("\nStatistik deskriptif data numerik:")
display(df.describe())

print("\nJumlah nilai unik pada kolom kategorikal:")
categorical_cols_initial = df.select_dtypes(include=['object']).columns
unique_values = df[categorical_cols_initial].nunique().sort_values(ascending=False)
display(unique_values)


# ===========================================================
# 5. DATA CLEANING
# ===========================================================

print("="*80)
print("DATA CLEANING")
print("="*80)

df_clean = df.copy()

# 5.1 Menghapus data duplikat
df_clean = df_clean.drop_duplicates()

# 5.2 Mengisi missing value
df_clean['children'] = df_clean['children'].fillna(0)
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean['agent'] = df_clean['agent'].fillna(0)
df_clean['company'] = df_clean['company'].fillna(0)

# 5.3 Menghapus data booking tanpa tamu
df_clean = df_clean[
    (df_clean['adults'] + df_clean['children'] + df_clean['babies']) > 0
]

# 5.4 Membuat variabel tambahan
df_clean['total_guests'] = (
    df_clean['adults'] + df_clean['children'] + df_clean['babies']
)

df_clean['total_stay_nights'] = (
    df_clean['stays_in_weekend_nights'] + df_clean['stays_in_week_nights']
)

print("Jumlah baris dan kolom sebelum cleaning:", df.shape)
print("Jumlah baris dan kolom setelah cleaning:", df_clean.shape)

print("\nMissing value setelah cleaning:")
missing_after = df_clean.isnull().sum().sort_values(ascending=False)
display(missing_after.head(10))

print("\nJumlah duplikasi setelah cleaning:")
print(df_clean.duplicated().sum())

display(df_clean.head())


# ===========================================================
# 6. DATA EXPLORATION
# ===========================================================

print("="*80)
print("DATA EXPLORATION")
print("="*80)

# 6.1 Distribusi status pembatalan
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='is_canceled')
plt.title('Distribusi Status Pembatalan Booking')
plt.xlabel('Status Pembatalan: 0 = Tidak Dibatalkan, 1 = Dibatalkan')
plt.ylabel('Jumlah Booking')
plt.tight_layout()
plt.show()

# 6.2 Distribusi tipe hotel
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='hotel')
plt.title('Distribusi Tipe Hotel')
plt.xlabel('Tipe Hotel')
plt.ylabel('Jumlah Booking')
plt.tight_layout()
plt.show()

# 6.3 Distribusi market segment
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df_clean,
    y='market_segment',
    order=df_clean['market_segment'].value_counts().index
)
plt.title('Distribusi Market Segment')
plt.xlabel('Jumlah Booking')
plt.ylabel('Market Segment')
plt.tight_layout()
plt.show()


# ===========================================================
# 7. HISTOGRAM
# ===========================================================

print("="*80)
print("HISTOGRAM")
print("="*80)

plt.figure(figsize=(8, 5))
sns.histplot(df_clean['lead_time'], bins=30, kde=True)
plt.title('Histogram Lead Time')
plt.xlabel('Lead Time')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df_clean['total_stay_nights'], bins=30, kde=True)
plt.title('Histogram Total Stay Nights')
plt.xlabel('Total Stay Nights')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df_clean['adr'], bins=30, kde=True)
plt.title('Histogram Average Daily Rate')
plt.xlabel('ADR')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()


# ===========================================================
# 8. BOXPLOT
# ===========================================================

print("="*80)
print("BOXPLOT")
print("="*80)

boxplot_cols = [
    'lead_time',
    'total_stay_nights',
    'total_guests',
    'adr',
    'total_of_special_requests'
]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean[boxplot_cols])
plt.title('Boxplot Variabel Numerik Utama')
plt.xlabel('Variabel')
plt.ylabel('Nilai')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df_clean, x='is_canceled', y='lead_time')
plt.title('Boxplot Lead Time Berdasarkan Status Pembatalan')
plt.xlabel('Status Pembatalan')
plt.ylabel('Lead Time')
plt.tight_layout()
plt.show()


# ===========================================================
# 9. HEATMAP MATRIX
# ===========================================================

print("="*80)
print("HEATMAP MATRIX")
print("="*80)

numerical_cols = [
    'is_canceled',
    'lead_time',
    'arrival_date_year',
    'arrival_date_week_number',
    'arrival_date_day_of_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'adults',
    'children',
    'babies',
    'is_repeated_guest',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'booking_changes',
    'days_in_waiting_list',
    'adr',
    'required_car_parking_spaces',
    'total_of_special_requests',
    'total_guests',
    'total_stay_nights'
]

corr_matrix = df_clean[numerical_cols].corr()

plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Heatmap Korelasi Variabel Numerik')
plt.tight_layout()
plt.show()


# ===========================================================
# 10. DETEKSI OUTLIER DENGAN IQR
# ===========================================================

print("="*80)
print("DETEKSI OUTLIER DENGAN IQR")
print("="*80)

outlier_summary = []

for col in boxplot_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = df_clean[
        (df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)
    ].shape[0]

    outlier_summary.append({
        'Variabel': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Batas Bawah': lower_bound,
        'Batas Atas': upper_bound,
        'Jumlah Outlier': outlier_count
    })

outlier_summary_df = pd.DataFrame(outlier_summary)
display(outlier_summary_df)

print("""
Catatan:
Outlier pada penelitian ini tidak langsung dihapus karena nilai ekstrem seperti lead time tinggi,
durasi menginap panjang, atau ADR tinggi masih dapat merepresentasikan perilaku pemesanan
pelanggan yang nyata dalam industri hotel.
""")


# ===========================================================
# 11. DATA TRANSFORMATION UNTUK APRIORI
# ===========================================================

print("="*80)
print("DATA TRANSFORMATION UNTUK APRIORI")
print("="*80)

df_apriori = df_clean.copy()

df_apriori['lead_time_category'] = pd.cut(
    df_apriori['lead_time'],
    bins=[-1, 7, 30, 90, 180, 1000],
    labels=['Very_Short', 'Short', 'Medium', 'Long', 'Very_Long']
)

df_apriori['stay_category'] = pd.cut(
    df_apriori['total_stay_nights'],
    bins=[-1, 1, 3, 7, 14, 100],
    labels=['One_Night', 'Short_Stay', 'Medium_Stay', 'Long_Stay', 'Very_Long_Stay']
)

df_apriori['guest_category'] = pd.cut(
    df_apriori['total_guests'],
    bins=[-1, 1, 2, 4, 100],
    labels=['Single_Guest', 'Couple', 'Small_Group', 'Large_Group']
)

df_apriori['special_request_category'] = pd.cut(
    df_apriori['total_of_special_requests'],
    bins=[-1, 0, 2, 5],
    labels=['No_Request', 'Few_Requests', 'Many_Requests']
)

df_apriori['parking_category'] = np.where(
    df_apriori['required_car_parking_spaces'] > 0,
    'Need_Parking',
    'No_Parking'
)

df_apriori['cancel_status'] = np.where(
    df_apriori['is_canceled'] == 1,
    'Canceled_Yes',
    'Canceled_No'
)

df_apriori['repeated_guest_status'] = np.where(
    df_apriori['is_repeated_guest'] == 1,
    'Repeated_Guest',
    'New_Guest'
)

category_cols = [
    'lead_time_category',
    'stay_category',
    'guest_category',
    'special_request_category'
]

df_apriori = df_apriori.dropna(subset=category_cols)

print("Jumlah data setelah transformasi kategori:", df_apriori.shape)

display(df_apriori[
    [
        'lead_time',
        'lead_time_category',
        'total_stay_nights',
        'stay_category',
        'total_guests',
        'guest_category',
        'total_of_special_requests',
        'special_request_category',
        'cancel_status'
    ]
].head())


# ===========================================================
# 12. DATA SELECTION UNTUK APRIORI
# ===========================================================

print("="*80)
print("DATA SELECTION UNTUK APRIORI")
print("="*80)

transaction_columns = [
    'hotel',
    'meal',
    'market_segment',
    'distribution_channel',
    'customer_type',
    'deposit_type',
    'reserved_room_type',
    'lead_time_category',
    'stay_category',
    'guest_category',
    'special_request_category',
    'parking_category',
    'cancel_status',
    'repeated_guest_status'
]

transaction_data = df_apriori[transaction_columns].astype(str)

print("Kolom yang digunakan untuk Apriori:")
print(transaction_columns)

print("\nContoh data yang akan dijadikan transaksi:")
display(transaction_data.head())


# ===========================================================
# 13. MEMBUAT DATA TRANSAKSI
# ===========================================================

print("="*80)
print("MEMBUAT DATA TRANSAKSI")
print("="*80)

transactions = []

for _, row in transaction_data.iterrows():
    transaction = []
    for col in transaction_columns:
        item = f"{col}_{row[col]}"
        transaction.append(item)
    transactions.append(transaction)

print("Jumlah transaksi:", len(transactions))
print("\nContoh transaksi pertama:")
print(transactions[0])


# ===========================================================
# 14. ONE-HOT ENCODING TRANSAKSI
# ===========================================================

print("="*80)
print("ONE-HOT ENCODING TRANSAKSI")
print("="*80)

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket_df = pd.DataFrame(te_array, columns=te.columns_)

print("Ukuran data transaksi setelah one-hot encoding:", basket_df.shape)
display(basket_df.head())
```

```python
# ===========================================================
# 15. ALGORITMA APRIORI
# ===========================================================

print("="*80)
print("ALGORITMA APRIORI")
print("="*80)

min_support_value = 0.08

frequent_itemsets = apriori(
    basket_df,
    min_support=min_support_value,
    use_colnames=True,
    max_len=4
)

if frequent_itemsets.empty:
    print("Frequent itemsets kosong pada min_support 0.08. Min_support diturunkan ke 0.05.")
    min_support_value = 0.05
    frequent_itemsets = apriori(
        basket_df,
        min_support=min_support_value,
        use_colnames=True,
        max_len=4
    )

if frequent_itemsets.empty:
    print("Frequent itemsets masih kosong pada min_support 0.05. Min_support diturunkan ke 0.03.")
    min_support_value = 0.03
    frequent_itemsets = apriori(
        basket_df,
        min_support=min_support_value,
        use_colnames=True,
        max_len=4
    )

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)

print("Min support yang digunakan:", min_support_value)
print("Jumlah frequent itemsets:", len(frequent_itemsets))

display(frequent_itemsets.head(30))


# ===========================================================
# 16. ASSOCIATION RULES
# ===========================================================

print("="*80)
print("ASSOCIATION RULES")
print("="*80)

min_confidence_value = 0.70

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=min_confidence_value
)

if rules.empty:
    print("Rules kosong pada confidence 0.70. Confidence diturunkan ke 0.60.")
    min_confidence_value = 0.60
    rules = association_rules(
        frequent_itemsets,
        metric="confidence",
        min_threshold=min_confidence_value
    )

if rules.empty:
    print("Rules masih kosong pada confidence 0.60. Confidence diturunkan ke 0.50.")
    min_confidence_value = 0.50
    rules = association_rules(
        frequent_itemsets,
        metric="confidence",
        min_threshold=min_confidence_value
    )

rules = rules.sort_values(
    by=['lift', 'confidence', 'support'],
    ascending=False
)

rules_display = rules.copy()
rules_display['antecedents'] = rules_display['antecedents'].apply(lambda x: ', '.join(list(x)))
rules_display['consequents'] = rules_display['consequents'].apply(lambda x: ', '.join(list(x)))

selected_rules = rules_display[
    [
        'antecedents',
        'consequents',
        'support',
        'confidence',
        'lift',
        'leverage',
        'conviction'
    ]
]

print("Min confidence yang digunakan:", min_confidence_value)
print("Jumlah association rules:", len(selected_rules))

print("\nTop 30 Association Rules berdasarkan lift:")
display(selected_rules.head(30))


# ===========================================================
# 17. RULES BERDASARKAN CONFIDENCE TERTINGGI
# ===========================================================

print("="*80)
print("RULES DENGAN CONFIDENCE TERTINGGI")
print("="*80)

top_confidence_rules = selected_rules.sort_values(
    by=['confidence', 'lift', 'support'],
    ascending=False
).head(20)

display(top_confidence_rules)

print("""
Bagian ini dapat digunakan untuk melihat aturan dengan tingkat ketepatan tertinggi.
Pada Apriori, confidence dapat dipahami sebagai tingkat keyakinan sebuah aturan.
""")


# ===========================================================
# 18. RULES TERKAIT PEMBATALAN BOOKING
# ===========================================================

print("="*80)
print("RULES TERKAIT PEMBATALAN BOOKING")
print("="*80)

cancel_rules = selected_rules[
    selected_rules['consequents'].str.contains('cancel_status_Canceled_Yes', case=False, na=False)
].sort_values(
    by=['confidence', 'lift', 'support'],
    ascending=False
)

not_cancel_rules = selected_rules[
    selected_rules['consequents'].str.contains('cancel_status_Canceled_No', case=False, na=False)
].sort_values(
    by=['confidence', 'lift', 'support'],
    ascending=False
)

print("Jumlah rules yang mengarah ke pembatalan booking:", len(cancel_rules))
display(cancel_rules.head(20))

print("Jumlah rules yang mengarah ke booking tidak dibatalkan:", len(not_cancel_rules))
display(not_cancel_rules.head(20))


# ===========================================================
# 19. EVALUASI KUALITAS RULES APRIORI
# ===========================================================

print("="*80)
print("EVALUASI KUALITAS RULES APRIORI")
print("="*80)

if len(selected_rules) > 0:
    rule_accuracy = selected_rules['confidence'].mean() * 100

    print("Jumlah rules:", len(selected_rules))
    print("Rata-rata Support:", round(selected_rules['support'].mean(), 4))
    print("Rata-rata Confidence:", round(selected_rules['confidence'].mean(), 4))
    print("Rata-rata Lift:", round(selected_rules['lift'].mean(), 4))

    print("\nSupport tertinggi:", round(selected_rules['support'].max(), 4))
    print("Confidence tertinggi:", round(selected_rules['confidence'].max(), 4))
    print("Lift tertinggi:", round(selected_rules['lift'].max(), 4))

    print("\nPendekatan tingkat ketepatan rules berdasarkan rata-rata confidence:")
    print(round(rule_accuracy, 2), "%")
else:
    rule_accuracy = np.nan
    print("Tidak ada association rules yang terbentuk.")


# ===========================================================
# 20. EVALUASI RULES PEMBATALAN BOOKING
# ===========================================================

print("="*80)
print("EVALUASI RULES PEMBATALAN BOOKING")
print("="*80)

if len(cancel_rules) > 0:
    cancel_rule_accuracy = cancel_rules['confidence'].mean() * 100

    print("Jumlah cancel rules:", len(cancel_rules))
    print("Rata-rata Support Cancel Rules:", round(cancel_rules['support'].mean(), 4))
    print("Rata-rata Confidence Cancel Rules:", round(cancel_rules['confidence'].mean(), 4))
    print("Rata-rata Lift Cancel Rules:", round(cancel_rules['lift'].mean(), 4))

    print("\nPendekatan tingkat ketepatan rules pembatalan berdasarkan confidence:")
    print(round(cancel_rule_accuracy, 2), "%")
else:
    cancel_rule_accuracy = np.nan
    print("Tidak ada rules yang mengarah ke pembatalan booking.")


# ===========================================================
# 21. TABEL RINGKASAN EVALUASI APRIORI
# ===========================================================

print("="*80)
print("TABEL RINGKASAN EVALUASI APRIORI")
print("="*80)

evaluation_summary = pd.DataFrame({
    'Evaluasi': [
        'Jumlah Frequent Itemsets',
        'Jumlah Association Rules',
        'Min Support',
        'Min Confidence',
        'Rata-rata Support',
        'Rata-rata Confidence',
        'Rata-rata Lift',
        'Support Tertinggi',
        'Confidence Tertinggi',
        'Lift Tertinggi',
        'Pendekatan Ketepatan Rules (%)',
        'Jumlah Cancel Rules',
        'Pendekatan Ketepatan Cancel Rules (%)'
    ],
    'Nilai': [
        len(frequent_itemsets),
        len(selected_rules),
        min_support_value,
        min_confidence_value,
        round(selected_rules['support'].mean(), 4) if len(selected_rules) > 0 else np.nan,
        round(selected_rules['confidence'].mean(), 4) if len(selected_rules) > 0 else np.nan,
        round(selected_rules['lift'].mean(), 4) if len(selected_rules) > 0 else np.nan,
        round(selected_rules['support'].max(), 4) if len(selected_rules) > 0 else np.nan,
        round(selected_rules['confidence'].max(), 4) if len(selected_rules) > 0 else np.nan,
        round(selected_rules['lift'].max(), 4) if len(selected_rules) > 0 else np.nan,
        round(rule_accuracy, 2) if len(selected_rules) > 0 else np.nan,
        len(cancel_rules),
        round(cancel_rule_accuracy, 2) if len(cancel_rules) > 0 else np.nan
    ]
})

display(evaluation_summary)


# ===========================================================
# 22. VISUALISASI HASIL APRIORI
# ===========================================================

print("="*80)
print("VISUALISASI HASIL APRIORI")
print("="*80)

top_lift_rules = selected_rules.sort_values(by='lift', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_lift_rules,
    x='lift',
    y='antecedents'
)
plt.title('Top 10 Association Rules Berdasarkan Lift')
plt.xlabel('Lift')
plt.ylabel('Antecedents')
plt.tight_layout()
plt.show()

top_conf_rules_plot = selected_rules.sort_values(by='confidence', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_conf_rules_plot,
    x='confidence',
    y='antecedents'
)
plt.title('Top 10 Association Rules Berdasarkan Confidence')
plt.xlabel('Confidence')
plt.ylabel('Antecedents')
plt.tight_layout()
plt.show()

# Scatter plot yang diperbaiki:
# hanya menampilkan 100 rules terbaik agar grafik tidak terlalu padat
top_rules_scatter = selected_rules.sort_values(
    by=['lift', 'confidence', 'support'],
    ascending=False
).head(100)

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=top_rules_scatter,
    x='support',
    y='confidence',
    size='lift',
    hue='lift',
    sizes=(50, 300),
    alpha=0.75
)
plt.title('Sebaran Top 100 Association Rules Berdasarkan Support, Confidence, dan Lift')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.tight_layout()
plt.show()

eval_plot = evaluation_summary[
    evaluation_summary['Evaluasi'].isin([
        'Rata-rata Support',
        'Rata-rata Confidence',
        'Rata-rata Lift'
    ])
].copy()

eval_plot['Nilai'] = eval_plot['Nilai'].astype(float)

plt.figure(figsize=(8, 5))
sns.barplot(data=eval_plot, x='Evaluasi', y='Nilai')
plt.title('Ringkasan Evaluasi Apriori')
plt.xlabel('Metrik')
plt.ylabel('Nilai')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()


# ===========================================================
# 23. INTERPRETASI OTOMATIS 10 RULES TERATAS
# ===========================================================

print("="*80)
print("INTERPRETASI OTOMATIS 10 RULES TERATAS")
print("="*80)

top_10_rules = selected_rules.sort_values(
    by=['lift', 'confidence', 'support'],
    ascending=False
).head(10)

for idx, row in top_10_rules.iterrows():
    print("\nRule")
    print("Jika pelanggan memiliki pola:")
    print(row['antecedents'])
    print("Maka cenderung berkaitan dengan:")
    print(row['consequents'])
    print("Support:", round(row['support'], 4))
    print("Confidence:", round(row['confidence'], 4))
    print("Lift:", round(row['lift'], 4))


# ===========================================================
# 24. KESIMPULAN OTOMATIS SINGKAT
# ===========================================================

print("="*80)
print("KESIMPULAN OTOMATIS SINGKAT")
print("="*80)

if len(selected_rules) > 0:
    best_rule = selected_rules.sort_values(
        by=['lift', 'confidence', 'support'],
        ascending=False
    ).iloc[0]

    print("""
Berdasarkan hasil penerapan algoritma Apriori, ditemukan sejumlah frequent itemsets
dan association rules yang menggambarkan pola perilaku pemesanan pelanggan hotel.
Evaluasi aturan dilakukan menggunakan support, confidence, dan lift.
""")

    print("Rule terbaik berdasarkan kombinasi lift, confidence, dan support:")
    print("Antecedents:", best_rule['antecedents'])
    print("Consequents:", best_rule['consequents'])
    print("Support:", round(best_rule['support'], 4))
    print("Confidence:", round(best_rule['confidence'], 4))
    print("Lift:", round(best_rule['lift'], 4))

    print("\nPendekatan tingkat ketepatan seluruh rules:", round(rule_accuracy, 2), "%")

    if len(cancel_rules) > 0:
        print("Pendekatan tingkat ketepatan cancel rules:", round(cancel_rule_accuracy, 2), "%")

print("""
Catatan laporan:
Istilah accuracy tidak digunakan secara langsung pada Apriori.
Nilai confidence digunakan sebagai pendekatan tingkat ketepatan aturan asosiasi.
""")


# ===========================================================
# 25. EXPORT HASIL KE EXCEL
# ===========================================================

print("="*80)
print("EXPORT HASIL KE EXCEL")
print("="*80)

output_file = "hasil_lengkap_apriori_hotel_booking.xlsx"

with pd.ExcelWriter(output_file) as writer:
    missing_before.to_excel(writer, sheet_name="Missing Before")
    missing_after.to_excel(writer, sheet_name="Missing After")
    outlier_summary_df.to_excel(writer, sheet_name="Outlier Summary", index=False)
    frequent_itemsets.to_excel(writer, sheet_name="Frequent Itemsets", index=False)
    selected_rules.to_excel(writer, sheet_name="Association Rules", index=False)
    top_confidence_rules.to_excel(writer, sheet_name="Top Confidence Rules", index=False)
    cancel_rules.to_excel(writer, sheet_name="Cancel Rules", index=False)
    not_cancel_rules.to_excel(writer, sheet_name="Not Cancel Rules", index=False)
    evaluation_summary.to_excel(writer, sheet_name="Evaluation Summary", index=False)

print("File berhasil disimpan:", output_file)

files.download(output_file)
```
